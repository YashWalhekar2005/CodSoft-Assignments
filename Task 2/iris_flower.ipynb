```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = pd.read_csv('Iris.csv')
iris.head()
```

```python
iris= iris.rename(columns={'SepalLengthCm':'Sepal_Length',
                           'SepalWidthCm':'Sepal_Width',
                           'PetalLengthCm':'Petal_Length',
                           'PetalWidthCm':'Petal_Width'})
```

```python
iris.head()
```

```python
iris.isnull().sum()
```

```python
iris ['Species'].value_counts()
```

```python
iris.describe()
```

```python
sns.FacetGrid(iris, hue="Species",height=6).map(plt.scatter,"Petal_Length","Sepal_Width").add_legend()
```

```python
sns.pairplot(iris[['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width','Species']], hue="Species",diag_kind='kde')
```

```python
data=iris.values
X=data[:,0:4]
Y=data[:,5]
```

```python
print(X.shape)
print(X)
```

```python
print(Y.shape)
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,Y, test_size=0.2)
```

```python
print(X_train.shape)
print(X_train)
```

```python
print(y_test.shape)
print(y_test)
```

```python
from sklearn.svm import SVC
model_svc=SVC()
model_svc.fit(X_train,y_train)
```

```python
prediction1 = model_svc.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,  prediction1))
```

```python
X_new = np.array([[3, 2, 1, 0.2], [4.9, 2.2, 3.8, 1.1], [5.3, 2.5, 4.6, 1.9]])

predicted_sizes = model_svc.predict(X_new)

# Output the predicted sizes
print(predicted_sizes)
```
